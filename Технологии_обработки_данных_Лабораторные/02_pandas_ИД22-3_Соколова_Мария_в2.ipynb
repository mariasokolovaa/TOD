5.1 При помощи объединения таблиц, создайте `DataFrame`, состоящий из четырех столбцов: `id`, `name`, `user_id`, `rating`. Рецепты, на которые не оставлен ни один отзыв, должны отсутствовать в полученной таблице. 5.2 При помощи объединения таблиц и группировок, создайте `DataFrame`, состоящий из трех столбцов: `recipe_id`, `name`, `review_count`, где столбец `review_count` содержит кол-во отзывов, оставленных на рецепт `recipe_id`. У рецептов, на которые не оставлен ни один отзыв, в столбце `review_count` должен быть указан 0. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и найдя строку, соответствующую этому рецепту, в полученном `DataFrame`.

```python
a=pd.DataFrame(reviews['recipe_id'].value_counts())
a['id']=a.index
a.index=[[i for i in range(a.shape[0])]]
a.columns = ['review_count','id']
df5_2=pd.merge(a,reviews)
df5_2=pd.merge(df5_2,recipes)
df5_2=df5_2[['recipe_id', 'name', 'review_count']]
recipes['recipe_id']=recipes['id']
s2=pd.merge(recipes,reviews,how='outer')
s2=s2[s2['rating'].isna()]
df5_2=pd.merge(df5_2,s2,how='outer')
df5_2=df5_2[['review_count', 'name', 'recipe_id']]
df5_2['review_count']=df5_2['review_count'].fillna(0)
df5_2
```
